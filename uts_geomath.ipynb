## 1.1 The Binomial Theorem    
### Teorema Binomial adalah suatu rumus dalam aljabar yang digunakan untuk menguraikan (a + b)‚Åø, yaitu pengangkatan dari suatu binomial (suatu ekspresi dengan dua suku) menjadi penjumlahan beberapa suku yang melibatkan koefisien binomial.  
       
### Rumusnya :   
### $(a + b)^n = \sum_{k=0}^{n} \binom{n}{k} a^{n-k} b^k$   
### atau dapat ditulis pula:   
### $(a + b)^n = \binom{n}{0}a^n+\binom{n}{1}a^{n-1}b+\binom{n}{2}a^{n-2}b^2+...+\binom{n}{n}b^n$   
     
### di mana $\binom{n}{k}$ adalah koefisien binomial, yang dapat dihitung dengan rumus:   
### $\frac{n!}{k!(n-k)!}$

```python
import numpy as np
import matplotlib.pyplot as plt

# Buat lingkaran
theta = np.linspace(0, 2*np.pi, 300)
circle_x = np.cos(theta)
circle_y = np.sin(theta)

# Tentukan titik sudut dan busur
angle = 1  # 1 radian
arc_x = np.cos(np.linspace(0, angle, 50))
arc_y = np.sin(np.linspace(0, angle, 50))

# Buat plot
fig, ax = plt.subplots(figsize=(6,6))
ax.plot(circle_x, circle_y, 'b', label="Lingkaran")
ax.plot([0, 1], [0, 0], 'r', linewidth=2, label="Radius r")  
ax.plot([0, arc_x[-1]], [0, arc_y[-1]], 'r', linewidth=2)  
ax.plot(arc_x, arc_y, 'g', linewidth=2, label="Busur r")  

ax.text(0.5, -0.1, "r", fontsize=14, color='r', ha='center')
ax.text((arc_x[-1]+0.1), arc_y[-1], "r", fontsize=14, color='g', ha='center')
ax.text(0.2, 0.2, "1 radian", fontsize=14, color='purple')

ax.set_xlim(-1.2, 1.2)
ax.set_ylim(-1.2, 1.2)
ax.set_xticks([])
ax.set_yticks([])
ax.set_aspect('equal')
ax.legend()
ax.set_title("Definisi Radian")

plt.show()
```

## SINE RULE
### Sine rule atau aturan sinus adalah hubungan dalam trigonometri yang digunakan pada segitiga, khususnya segitiga sembarang (bukan hanya siku-siku). Aturan ini menyatakan bahwa perbandingan antara panjang sisi segitiga dan sinus sudut yang berhadapan dengannya adalah sama untuk ketiga sisi. Secara matematis ditulis sebagai:
$$
\frac{a}{sin A}=\frac{b}{sin B}=\frac{c}{sin C}
$$
### di mana a, b, C adalah panjang sisi-sisi segitiga, dan A, B, C sudut-sudut yang berhadapan langsung dengan sisi-sisi tersebut. Aturan ini sangat berguna untuk menghitung sisi atau sudut segitiga jika diketahui satu pasang sisi dan sudut yang berhadapan serta satu elemen lain dari segitiga. Aturan sinus berlaku untuk semua jenis segitiga, dan sering digunakan dalam pemecahan masalah geometri, navigasi, serta fisika vektor.

## COSINE RULE
### Cosine rule atau aturan kosinus adalah rumus dalam trigonometri yang digunakan untuk menghitung panjang sisi atau besar sudut pada segitiga, terutama saat tidak bisa langsung menggunakan aturan sinus. Rumusnya:
$$
c^2=a^2+b^2-2ab cos C  
$$
### di mana a, b, dan c adalah panjang sisi-sisi segitiga, dan C adalah sudut yang berhadapan dengan sisi c. Rumus ini merupakan perluasan dari Teorema Pythagoras, dan berlaku untuk semua segitiga, baik siku-siku maupun tidak. Jika sudut C adalah 90¬∞, maka cos ùê∂ = 0 cos C=0, dan rumusnya menjadi Teorema Pythagoras. Aturan kosinus berguna untuk menghitung sisi jika dua sisi dan sudut di antaranya diketahui, atau menghitung sudut jika ketiga sisi diketahui.

# **3. Complex Number**

### Dalam dunia matematika, terdapat banyak jenis bilangan yang digunakan untuk menyelesaikan berbagai jenis persoalan. Salah satu jenis bilangan yang sangat penting, namun tidak termasuk dalam bilangan real, adalah bilangan kompleks. Bilangan ini diperkenalkan sebagai solusi dari persamaan yang tidak memiliki akar real, seperti $x^2 + 1 = 0$.
### Bilangan kompleks adalah bilangan yang memiliki bagian real dan bagian imajiner. Bilangan kompleks sering ditulis dalam bentuk:    
### $z = a+ bi$     
### Di mana:    
### $\cdot$ a adalah bagian riil,    
### $\cdot$ b adalah bagian imajiner, dan    
### $\cdot$ bùëñ adalah bilangan imajiner yang memenuhi $i^2=‚àí1$.
### Bilangan kompleks memperluas sistem bilangan real dan memungkinkan kita untuk menyelesaikan persamaan-persamaan yang tidak memiliki solusi dalam bilangan real.

### Misalnya ‚àí1+j2 dan ‚àí1‚àíj2

### Namun, dalam bidang teknik, \( i \) sudah digunakan untuk melambangkan arus listrik, sehingga untuk menghindari kebingungan, digunakanlah huruf berikutnya dalam alfabet, yaitu \( j \), untuk mewakili: $$\sqrt{-1}$$


## 2. Addition and Substraction of Complex Numbers (Penjumlahan dan Pengurangan)

### Dua bilangan kompleks dijumlahkan atau dikurangkan dengan cara menjumlahkan atau mengurangkan bagian real dan bagian imajinernya secara terpisah.

### Sebagai contoh, jika:
### $Z_1 = a + jb$ dan $Z_2 = c + jd$
### maka:

### $Z_1 + Z_2 = (a+jb)+(c+jd)=(a+c)+j(b+d)$

### dan

### $Z_1 - Z_2 = (a+jb)-(c+jd)=(a-c)+j(b-d)$

## 3. Multiplication and Division of Complex Numbers (Perkalian dan Pembagian)

### Perkalian bilangan kompleks dilakukan dengan menganggap semua komponen sebagai bilangan real terlebih dahulu, lalu menyederhanakannya menggunakan fakta bahwa ùëó2 = ‚àí1

### Konjugasi kompleks dari suatu bilangan kompleks diperoleh dengan mengubah tanda bagian imajinernya. Jadi, konjugasi kompleks dari ùëé + ùëóùëè adalag ùëé ‚àí ùëóùëè. Hasil perkalian antara suatu bilangan kompleks dan konjugasinya selalu menghasilkan bilangan real.

### Contoh:
$$
\begin{aligned}
(3 + j4)(3 - j4) &= 9 - j12 + j12 - j^2 \cdot 16 \\
&= 9 + 16 \\
&= 25
\end{aligned}
$$

### Pembagian bilangan kompleks dilakukan dengan mengalikan pembilang dan penyebut dengan konjugasi kompleks dari penyebut. Tujuannya adalah untuk menghilangkan bagian imajiner di penyebut, sehingga hasil akhirnya bisa disederhanakan.

## 4.1 Gradient 
### Gradien dalam matematika, khususnya di grafik fungsi atau geometri analitik, adalah ukuran kemiringan sebuah garis.    
$$
\frac{\Delta y}{\Delta x} \tag{1}
$$

### atau   


$$
\frac{y_2 - y_1}{x_2 - x_1} \tag{2}
$$


### Diferensial adalah pengembangan konsep gradien ke fungsi yang tidak lurus (kurva). kalau gradien kita pakai untuk garis lurus maka diferensial dipakai untuk menentukan kemiringan di titik tertentu dari suatu kurva. dalam artian Dfernsial ini adalah bentuk gradien lokal dari fungsi yang melengkung (non-linier).

### Jika $ y = f(x) $ maka turunan pertama didefinisikan sebagai:
$$
\frac{dy}{dx} = \lim_{\Delta x \to 0} \frac{f(x + \Delta x) - f(x)}{\Delta x} \tag{3}
$$

### Atau dalam notasi lain:

$$
f'(x) = \lim_{h \to 0} \frac{f(x + h) - f(x)}{h} \tag{4}
$$

## 4.2 Differentiation Rules (Aturan Turunan)
### Differentiation adalah proses mencari turunan dari suatu fungsi. Turunan menunjukkan laju perubahan atau gradien fungsi di suatu titik

### 1. Basic Rules (Aturan Dasar)

### | Fungsi $ f(x) $ | Turunan $ f'(x) $ | Keterangan |
### | Konstanta $ c $ | 0 | Turunan konstanta |
### $ x^n $ | $ n x^{n-1} $ | Aturan pangkat (Power Rule) |
### Contoh:
- $ \frac{d}{dx} (5) = 0 $
- $ \frac{d}{dx} (x^3) = 3x^2 $


### 2. Constant Multiple Rule (Aturan Konstanta)

### Jika fungsi dikali konstanta:
$$
\frac{d}{dx}[c \cdot f(x)] = c \cdot f'(x) \tag{7}
$$

### Contoh:
$
\frac{d}{dx}(7x^2) = 7 \times (2x) = 14x
$

### 3. Sum and Difference Rule (Penjumlahan dan Pengurangan)

### Jika dua fungsi dijumlah/dikurang:
$$
\frac{d}{dx}[f(x) \pm g(x)] = f'(x) \pm g'(x) \tag{8}
$$

### Contoh:
$
\frac{d}{dx}(x^2 + 3x) = 2x + 3
$

### 4. Product Rule (Aturan Perkalian)

### Jika dua fungsi dikali:
$$
\frac{d}{dx}[f(x)g(x)] = f'(x)g(x) + f(x)g'(x) \tag{9}
$$

### Contoh:
### Misal $ f(x) = x^2 $ dan $ g(x) = \sin(x) $,   
$
\frac{d}{dx}[x^2 \sin(x)] = 2x \sin(x) + x^2 \cos(x)
$


### 5. Quotient Rule (Aturan Pembagian)

### Jika dua fungsi dibagi:
$$
\frac{d}{dx}\left(\frac{f(x)}{g(x)}\right) = \frac{f'(x)g(x) - f(x)g'(x)}{[g(x)]^2} \tag{10}
$$

**Contoh:**
### Misal $ f(x) = x $, $ g(x) = x+1 $,   
$
\frac{d}{dx}\left(\frac{x}{x+1}\right) = \frac{(1)(x+1) - (x)(1)}{(x+1)^2} = \frac{1}{(x+1)^2}
$


### 6. Chain Rule (Aturan Rantai)

### Jika fungsi di dalam fungsi (fungsi komposit):
$$
\frac{d}{dx}[f(g(x))] = f'(g(x)) \times g'(x) \tag{11}
$$

**Contoh:**
### Misal $ f(x) = (3x+2)^5 $,   
$
\frac{d}{dx}[(3x+2)^5] = 5(3x+2)^4 \times 3 = 15(3x+2)^4
$

## 4.3 Turunan Parsial (Partial Derivatives)

### Turunan parsial adalah turunan dari fungsi banyak variabel terhadap satu variabel saja, sedangkan variabel lainnya dianggap konstan.Biasanya digunakan untuk fungsi seperti:
$$
f(x, y), \quad f(x, y, z)
$$

### Notasi Turunan Parsial

### Kalau fungsi $ f(x, y) $, maka:

### - Turunan parsial terhadap $ x $:
$$
\frac{\partial f}{\partial x} \tag{12}
$$
### - Turunan parsial terhadap $ y $:
$$
\frac{\partial f}{\partial y} \tag{13}
$$

### Catatan: Simbol $ \partial $ (baca: "d-partial") digunakan untuk membedakan dari turunan biasa.
### Cara Menghitung

### Saat menghitung turunan parsial:
### - Variabel yang diturunkan dianggap berubah.
### - Variabel lain dianggap konstan (seperti angka biasa).

## 4.4 Maxima or Minima
### Dalam kalkulus multivariabel, kita sering kali tertarik untuk menemukan nilai maksimum atau minimum dari suatu fungsi yang melibatkan lebih dari satu variabel. Fungsi tersebut disebut **fungsi multivariabel**. Misalnya, fungsi dua variabel $z = f(x, y)$ sering ditemukan dalam geofisika untuk menggambarkan topografi atau distribusi medan gravitasi.    
### Langkah-langkah Mencari Maxima dan Minima

### 1. Hitung turunan pertama $ f'(x) $.
### 2. Cari titik kritis:
  $$ f'(x) = 0  \tag{17} $$
###  - Atau saat turunan tidak terdefinisi.
### 3. Gunakan uji turunan kedua:
###  - Jika $ f''(x) > 0 $, maka ada minimum lokal.
###  - Jika $ f''(x) < 0 $, maka ada maksimum lokal.
 ###  - Jika $ f''(x) = 0 $, perlu analisa lebih lanjut.    

 ### Titik Stasioner
### Titik stasioner pada fungsi multivariabel adalah titik di mana turunan pertama terhadap semua variabel adalah nol. Secara matematis, ini berarti:
$$
\frac{\partial z}{\partial x} = 0 \quad \text{dan} \quad \frac{\partial z}{\partial y} = 0 \tag{4.12}
$$
### Titik tersebut disebut **titik stasioner** karena nilai fungsi di titik ini tidak berubah seiring perubahan kecil pada x dan y. Namun, titik ini belum tentu merupakan titik maksimum atau minimum.

### Jenis Titik Stasioner
### Untuk mengetahui apakah titik stasioner adalah maksimum atau minimum, kita menggunakan turunan kedua atau uji determinan dari matriks Hessian. Hessian adalah matriks yang berisi turunan kedua dari fungsi terhadap setiap pasangan variabel. Berdasarkan matriks Hessian, kita dapat menentukan jenis titik stasioner:
### - Jika determinannya positif dan turunan kedua terhadap x $\frac{\partial^2 z}{\partial x^2}$ juga positif, maka titik tersebut adalah **minimum lokal**.
### - Jika determinannya positif dan turunan kedua terhadap x $\frac{\partial^2 z}{\partial x^2}$ negatif, maka titik tersebut adalah **maksimum lokal**.
### - Jika determinannya negatif, maka titik tersebut adalah **titik sadle** (saddle point), yang berarti titik tersebut adalah maksimum di satu arah dan minimum di arah lain.

### Catatan Tambahan

### - Untuk fungsi banyak variabel (seperti \( f(x, y) \)), perlu menggunakan turunan parsial dan matriks Hessian.
### - Jika domain fungsi dibatasi, perlu juga cek nilai fungsi di batas domain.

### Ringkasan

### - Maxima ‚Üí puncak fungsi.
### - Minima ‚Üí lembah fungsi.
### - Cari critical points dengan turunan pertama.
### - Gunakan turunan kedua untuk menentukan jenis titik tersebut.

### Integrasi adalah konsep dasar dalam kalkulus yang merupakan kebalikan dari diferensiasi. Sederhananya, integrasi digunakan untuk menghitung "total" atau "luas" yang tercakup di bawah suatu kurva. ### Secara matematis, integrasi dapat dianggap sebagai proses mencari fungsi asal dari suatu fungsi turunan.

### Solusi umum dari integral dalam bentuk $\int ax^n dx$ di mana a dan n adalah konstanta, diberikan oleh
$$\int ax^n dx = \frac {ax^{n+1}}{n+1} + c  \tag {5.1} $$
